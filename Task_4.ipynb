```python
#Thêm các thư viện cần thiết
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
#Đọc file data
df = pd.read_csv('/content/drive/My Drive/Data/Final/stock.csv')
df
```

```python
#Scale dữ liệu
scaler = MinMaxScaler()

df_sc = df.copy()
df_sc.iloc[:, 1:] = scaler.fit_transform(df_sc.iloc[:, 1:])

df_sc
```

##**Stock price prediction: AAA**

```python
#Tìm index của 2022 để test -> index từ 431
df_test = df_sc[df_sc['Unnamed: 0'].str.contains("2022")]['AAA']
df_test
```

```python
#Lấy price của 2020, 2021 để train, Lấy price của 2022 để test 
data_AAA = df_sc['AAA'].values
data_train = data_AAA[:431]
data_test = data_AAA[431:]
```

```python
# SES
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

history = data_train
predictions = []
for new_data in data_test:
  model = SimpleExpSmoothing(history)
  model_fit = model.fit()
  prediction = model_fit.forecast(steps=1)    
  predictions.append(prediction[0])
  history = np.append(history, new_data)

print(predictions)
```

```python
df_tmp = df_sc.copy()
line = np.append(data_train, predictions)
df_tmp['AAA'] = line
df_tmp
```

```python
lines = scaler.inverse_transform(df_tmp.iloc[:, 1:])
lines_AAA = lines[:, 0]
prediction_line = lines_AAA[431:]
prediction_line
```

```python
plt.figure(figsize=(15, 5))


plt.plot(range(0, 530), df['AAA'].values, 'b-', label = 'Actual data')
plt.plot(range(431, 530), prediction_line, 'orange', label = 'Predicted data')

plt.title('Simple Exponential Smoothing Prediction')
plt.xlabel('Dates')
plt.ylabel('Stock price')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
